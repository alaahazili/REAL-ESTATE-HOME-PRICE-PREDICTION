```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
df = pd.read_csv('properties.csv')
df.head()
```

```python
df.shape
```

```python
df1 = df.drop('url',axis=1)
df1.head()
```

### Cleaning Data

```python
missing_values=df1.isnull().sum()
print(missing_values)
```

```python
df2=df1.dropna()
missing_values2=df2.isnull().sum()
missing_values2
```

```python
#clean bedrooms ,bathrooms columns (extract numerical data)
df2.loc[ : , 'Num_Bedrooms'] = df2['bedrooms'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))))
df2
```

```python
df2.loc[ : , 'Num_Bathrooms'] = df2['bathrooms'].apply(lambda x : int(''.join(filter(str.isdigit,str(x)))))
df2
```

```python
df2=df2.drop(['bedrooms','bathrooms'],axis=1)
df2
```

```python
df2['Num_Bathrooms'].unique()
```

### Clean Type , Location

```python
df2.location.value_counts()
```

```python
# Convert to lowercase and strip leading/trailing spaces
df2['location']=df2['location'].str.lower().str.strip()
df2['type']=df2['type'].str.lower().str.strip()

location_mapping= {
    'hay riyad': 'hay riad',
    'souissi ': 'souissi',
    'indéfini': 'unknown',
    'autre': 'unknown',
}
df2['location']=df2['location'].replace(location_mapping)
df2.fillna('unkown',inplace=True)
```

```python
df2
```

```python
len(df2.type.value_counts())
```

```python
len(df2.location.value_counts())
```

### Clean Area

```python
df2.area.unique()
```

```python
# Remove thin spaces and trailing spaces
df2['area'] = df2['area'].str.replace('\u202f', '', regex=False).str.strip()
```

```python
df2.area.unique()
```

```python
# Convert 'area' to numeric, coercing errors to NaN
df2.loc[ : , 'area'] = pd.to_numeric(df2['area'], errors='coerce')

# Fill NaN values with 0
df2.loc[ : , 'area'] = df2['area'].fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2['area_scaled'] = scaler.fit_transform(df2[['area']])
df2
```

```python
df2.area_scaled.describe()
```

```python
df2 = df2.drop('area', axis=1)
```

## Clean Price

```python
print(df2['price'].describe())
```

```python
Q1 = df2['price'].quantile(0.25)
Q3 = df2['price'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
df2 = df2[df2['price'] <= upper_bound]
```

```python
df2.loc[ : , 'price_log'] = np.log(df2['price'] + 1)  # +1 to avoid log(0)
df2
```

```python
df2['price_log'].describe()
```

```python
df2= df2.drop('price',axis=1)
```

```python
df3 = pd.get_dummies(df2, columns=['location', 'type'], drop_first=True)
df3
```

```python
# Clean column names (replace spaces with underscores)
df3.columns = df3.columns.str.replace(' ', '_')
```

```python
bool_cols = df3.select_dtypes(include='bool').columns
df3[bool_cols] = df3[bool_cols].astype(int)
df3
```

### Training Model 

```python
X = df3.drop('price_log',axis='columns')
```

```python
y = df3.price_log
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
regModel = LinearRegression()
regModel.fit(X_train, y_train)
regModel.fit(X_test, y_test)
```

```python
regModel.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import pandas as pd

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': make_pipeline(StandardScaler(), LinearRegression()),
            'params': {
                'linearregression__fit_intercept': [True, False],
                'linearregression__positive': [False]  # Optional parameter
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error'],
                'splitter': ['best', 'random'],
                'max_depth': [3, 5, 7],  # Reduce depth to limit complexity
                'min_samples_split': [5, 10],  # Require more samples to split
                'min_samples_leaf': [2, 4]  # Require more samples per leaf
            }
        },
        'random_forest' : {
            'model': RandomForestRegressor(),
            'params': {
                'n_estimators': [50, 100],
                'max_depth': [5, 7],
                'min_samples_split': [5, 10]
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    
    for algo_name, config in algos.items():
        gs = GridSearchCV(
            estimator=config['model'],
            param_grid=config['params'],
            cv=cv,
            return_train_score=False
        )
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
find_best_model_using_gridsearchcv(X,y)
```
